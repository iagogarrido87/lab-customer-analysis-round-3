```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/Users/iagogarrido/Documents/ironhack/unit1/lab-customer-analysis-round-3/files_for_lab/csv_files/marketing_customer_analysis.csv')
```

```python
#show data frame info
print(df.info())
```

```python
#Describe DataFrame.
print(df.describe())
```

```python
#Show a plot of the total number of responses
plt.figure(figsize=(8,6))
sns.countplot(x='Response', data=df)
```

```python
#Show a plot of the response rate by the sales channel.
df["Response"] = df["Response"].map({"Yes": 1, "No": 0})
plt.figure(figsize=(8,6))
sns.barplot(x='Sales Channel', y='Response', data=df)
```

```python
# Show a plot of the response rate by the total claim amount

plt.figure(figsize=(8,6))
sns.boxplot(x='Response', y='Total Claim Amount', data=df)
```

```python
# Show a plot of the response rate by income.
plt.figure(figsize=(8,6))
sns.boxplot(x='Response', y='Income', data=df)
```
